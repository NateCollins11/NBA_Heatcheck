```python
import math
newly_subbed_players = []
newly_subbed_makes_and_attempts = {"makes": 0, "attempts": 0}

for row in df.iterrows():
    playdata = row[1]
    if type(playdata.EnterGame) == str:
        
        ncode = name_with_code_to_ncode(playdata.EnterGame)

        newly_subbed_players.append(ncode)
#         print("adding " + ncode + " to newly_subbed_players list")
        
    if type(playdata.Shooter) == str:
        ncode = name_with_code_to_ncode(playdata.Shooter)
        if ncode in newly_subbed_players:
#             print(ncode + " takes a shot!")
            newly_subbed_makes_and_attempts["attempts"] += 1
            newly_subbed_players.remove(ncode)
            if playdata.ShotOutcome == "make":
                newly_subbed_makes_and_attempts["makes"] += 1
#             print("removing him from the newly subbed list")
            

print(newly_subbed_makes_and_attempts["makes"] / newly_subbed_makes_and_attempts["attempts"] * 100)
```
