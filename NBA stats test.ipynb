```python
pdf = pd.read_csv('all_seasons.csv')

# pdf['player_name'].map(lambda x: print(x))

pdf['player_name'] = pdf['player_name'].map(lambda x: 'ScrewYouNene' if (len(x.split()) < 2) else x.split()[1].lower()[:5] + x.split()[0].lower()[:2])
pdf.drop_duplicates(subset ="player_name", keep ='first', inplace = True)

pdf2018 = pdf[pdf['player_name'] == 'mooreet']
pdf2018.head(500)

# pdf

# pdf2018['player_name'].map(lambda x: print(x.split()[1].lower()[:5] + x.split()[0].lower()[:2]))
```

```python
player_with_fgperc_array = []


for player in df.Shooter.unique():
    
    player_shots = df[df["Shooter"] == player]

    if len(player_shots['ShotOutcome'].value_counts()) >= 2:
        player_ratio = player_shots['ShotOutcome'].value_counts().make / (player_shots['ShotOutcome'].value_counts().make + player_shots['ShotOutcome'].value_counts().miss)

#         print(player.split('- ')[1].split('0')[0], str(round(player_ratio * 100, 2)))

        new_name = player.split('- ')[1].split('0')[0]
        
        player_height = pdf.loc[pdf["player_name"] == new_name, "player_height"].tolist()
        playerheight = player_height[0] if (len(player_height) == 1) else "no value"
        
#         print(pdf.loc[pdf["player_name"] == new_name, "player_height"].tolist())
        if playerheight != "no value":
            player_with_fgperc_array.append([new_name, round(player_ratio * 100, 2), playerheight])
    
name_fgp_height = pd.DataFrame(player_with_fgperc_array, columns = ["name", "fg%", "height"])

name_fgp_height.head()
        
        
```

```python
import matplotlib as mpl

nfh = name_fgp_height.copy()

nfh["height"] = nfh["height"].map(lambda x: x * 0.3937)

bins = pd.IntervalIndex.from_tuples([(0, 70), (70, 72), (72, 74), (74, 76), (76, 78), (78, 80), (80, 82), (82, 84), (84, 86), (86, 100)])



groups = nfh.groupby(pd.cut(nfh["height"], bins)).mean()



groups.plot(kind = "bar", x = "height", y = "fg%")


```
